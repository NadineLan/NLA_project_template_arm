## Libraries and settings

```python
# Load Python libraries
import os
import pandas as pd
import matplotlib.pyplot as plt

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")

# Show current working directory
print(os.getcwd())
```

## Import data

```python
## Read data
df = pd.read_csv("./data/test.csv", sep=",")
df.head()
```

```python
x = 500
y = 300
plt.figure(figsize=(x/100, y/100))
plt.plot(df['A'], df['B'], marker='o', linestyle='-', color='blue')
plt.xlabel('A')
plt.ylabel('B')
plt.title('A vs B')
plt.grid(True)
plt.tight_layout()
plt.show()
            
```
