```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# LOAD DATA

```python
df = pd.read_csv("Davos_combined_cleaned.csv")
print("Daten erfolgreich geladen.")
```

# FILTER DATA 

```python
df_analysis = df[df['type'].isin(['ferienwohnung', 'hotel'])].copy()
num_rows = len(df_analysis)
print(f"\nNumber of entries for visualization: {num_rows}")
```

# CREATE SCATTER PLOT

```python
type_colors = {'ferienwohnung': 'blue', 'hotel': 'orange'}
unique_types = df_analysis['type'].unique()


plt.figure(figsize=(14, 8), dpi=300)

# Scatterplot: X=Distance Center, Y=Distance Ski Lift, Hue=Type, Size=Price
scatter_dist = sns.scatterplot(
    data=df_analysis,
    x='distance_zentrum_meters',
    y='distance_skilift_meters',
    hue='type',
    size='price',           # Price controls the size of the circle
    hue_order=['ferienwohnung', 'hotel'], 
    sizes=(50, 400),        # Range in which price sizes are scaled
    palette=type_colors, 
    alpha=.8,
    edgecolor='black',
    linewidth=0.5
)

#5. SET AXIS LIMITS
LIMIT_SKILIFT = 2000        # Y-axis limit
LIMIT_ZENTRUM = 3000        # X-axis limit

plt.xlim(0, LIMIT_ZENTRUM)
plt.ylim(0, LIMIT_SKILIFT)

plt.title('Distance from Center vs. Distance from Ski Lift, Colored by Type and Sized by Price', fontsize=16)
plt.xlabel('Distance to Center (meters)', fontsize=14)
plt.ylabel('Distance to Ski Lift (meters)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
```
